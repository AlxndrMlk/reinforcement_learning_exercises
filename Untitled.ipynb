```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
actual_rewards = [
    0, 4, 4, 4, 4, 4, 4, 2, 2, 2, 2, 4, 0, 4, 0, 4, 0, 4, 0
]
```

```python
def get_estimates(rewards, alpha):
    
    estimates = []
    
    alpha_is_func = 1 if type(alpha) is not float else 0
    
    for t, r in enumerate(rewards):
        
        try:
            q_t = estimates[-1]
        except IndexError:
            q_t = 0
            
        if alpha_is_func:
            alpha = alpha_func(t)
#             print(t, r, alpha)
            
        estimates.append(q_t + alpha*(r - q_t))
        
    return estimates
```

```python
def alpha_func(t):
    try:
        return 1 / (t)
    except ZeroDivisionError:
        return 0

alphas = [1., 1/2, 1/8, alpha_func]
```

```python
plt.figure(figsize=(12, 9))

for i, a in enumerate(alphas):
    
    estimates = get_estimates(actual_rewards, a)
    
    plt.subplot(2, 2, i + 1)
    
    plt.plot(actual_rewards, label='Actual rewards')
    plt.plot(estimates, label='Estimates', color='r', alpha=.7)
    plt.title(a)
    
```
